```python
ratings_equals_5_df = ratings_df[ratings_df.rating == 5]
```

```python
movieid_ratings_equals_5_df = pd.DataFrame(ratings_equals_5_df['movieId']\
                                           .value_counts())
```

```python
max_number_5_rate = max(ratings_equals_5_df['movieId'].value_counts())
```

```python
best_movieId = movieid_ratings_equals_5_df[movieid_ratings_equals_5_df\
                                           .movieId == max_number_5_rate]\
.index[0]

best_movie = movielens_df[movielens_df.movieId == best_movieId]['title'].\
to_string(index = False)

best_movie
```
